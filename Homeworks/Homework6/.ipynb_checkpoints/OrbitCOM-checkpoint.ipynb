```python
# import modules
import numpy as np
import astropy.units as u
from astropy.constants import G

# import plotting modules
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline

# my modules
from ReadFile import Read
# Step 1: modify CenterOfMass so that COM_P now takes a parameter specifying 
# by how much to decrease RMAX instead of a factor of 2
from CenterOfMass2 import CenterOfMass
```

```python
def OrbitCOM(galaxy, start, end, n):
    """function that loops over all the desired snapshots to compute the COM pos and vel as a function of time.
    inputs:
        galaxy -- (string) the name of the galaxy (e.g. "MW" for Milky Way)
        start -- (integer) the number of the first snapshot file to be read in 
        end -- (integer) the number of the last snapshot file to be read in 
        n -- (integer) the interval over which the COM is returned
        
    outputs: 
        Orbit_galaxyname.txt -- (txt file) a file that contains the results of the function:
            the COM positions, velocity vectors, and time of a given galaxy. 
    """
    
    # compose the filename for output
    fileout = f"Orbit_{galaxy}.txt"
    
    #  set tolerance and VolDec for calculating COM_P in CenterOfMass
    # for M33 that is stripped more, use different values for VolDec
    if galaxy == "M33":
        delta = 0.1
        volDec = 4 #use higher volume decrease bc later in time, M33 is stripped more. 
    else: 
        delta = 0.1
        volDec = 2 
    
    # generate the snapshot id sequence 
    # it is always a good idea to also check if the input is eligible (not required)
    snap_ids = np.arange(start,end,n)

    #check if input is empty. if it is, the code stops. 
    if len(snap_ids) == 0:
        
        print("this array is empty.")
        
        return 
    
    # initialize the array for orbital info: t, x, y, z, vx, vy, vz of COM
    orbit = np.zeros([len(snap_ids),7]) #this array has the same number of rows at snap_ids, and 7 columns to
    #store time, positon components, and velocity components
    
    # a for loop 
    for i,snap_id in enumerate(snap_ids): # loop over files
        
        # compose the data filename (be careful about the folder)
        ilbl = '000' + str(snap_id) #add a string of the filenumber to the value '000' (snap number 0)
        ilbl = ilbl[-3:] #remove all but the last 3 digits
        #filename = "%s_"%(galaxy) + ilbl + '.txt'
        filename=f"{galaxy}/{galaxy}_" + ilbl + ".txt" #having to store the filename differently bc they're in 
        #a different folder
        
        # Initialize an instance of CenterOfMass class, using disk particles
        COM = CenterOfMass(filename,2)

        # Store the COM pos and vel. Remember that now COM_P required VolDec
        COM_p = COM.COM_P(delta,volDec)
        COM_v = COM.COM_V(COM_p[0],COM_p[1],COM_p[2])

        #storing the values of the position components
        com_x = COM_p[0]
        com_y = COM_p[1]
        com_z = COM_p[2]

        #storing velocity components
        com_vx = COM_v[0]
        com_vy = COM_v[1]
        com_vz = COM_v[2]
        
        # store the time, pos, vel in ith element of  the orbit array, without units (use ().value)
        # note that you can store 
        # a[i] = var1, *tuple(array1)
        orbit[i][0] = ((COM.time).value/1000) #store time in GYR
        orbit[i][1] = (com_x).value
        orbit[i][2] = (com_y).value 
        orbit[i][3] = (com_z).value 
        orbit[i][4] = (com_vx).value
        orbit[i][5] = (com_vy).value
        orbit[i][6] = (com_vz).value
    
        # print snap_id to see the progress
        print(snap_id)
        
    # write the data to a file
    # we do this because we don't want to have to repeat this process 
    # this code should only have to be called once per galaxy.
    np.savetxt(fileout, orbit, fmt = "%11.3f"*7, comments='#',
               header="{:>10s}{:>11s}{:>11s}{:>11s}{:>11s}{:>11s}{:>11s}"\
                      .format('t', 'x', 'y', 'z', 'vx', 'vy', 'vz'))
    return 
```

```python
# Recover the orbits and generate the COM files for each galaxy
# read in 800 snapshots in intervals of n=5
# Note: This might take a little while - test your code with a smaller number of snapshots first! 
OrbitCOM("MW",0,800,5) 
OrbitCOM("M31",0,800,5)
OrbitCOM("M33",0,800,5)
```

```python
# Read in the data files for the orbits of each galaxy that you just created
# headers:  t, x, y, z, vx, vy, vz
# using np.genfromtxt
MW_data = np.genfromtxt("Orbit_MW.txt",dtype = None, names = True) 
M31_data = np.genfromtxt("Orbit_M31.txt",dtype = None, names = True) 
M33_data = np.genfromtxt("Orbit_M33.txt",dtype = None, names = True) 
print(MW_data) #test
```

```python
# function to compute the magnitude of the difference between two vectors 
# You can use this function to return both the relative position and relative velocity for two 
# galaxies over the entire orbit 

def magnitude(v1, v2):
    """function to compute the magnitude of the difference between 2 vectors
    inputs:
        v1 -- (array) the first vector (3 components)
        v2 -- (array) the second vector (3 components)
        
    outputs: 
        mag -- (float) the magnitude of the difference between the two vectors
    """
    #getting the difference. the difference is an array at this point. 
    diff_1 = v1[0] - v2[0] 
    diff_2 = v1[1] - v2[1] 
    diff_3 = v1[2] - v2[2] 
    
    #computing the mag: (x^2 + y^2 + z^2)
    mag = np.sqrt(diff_1**2 + diff_2**2 + diff_3**2) 
    
    return mag
```

```python
#test to make sure magnitude works 
v1 = np.array([1,1,1])
v2 = np.array([2,2,2])

print(v1)
print(v2)

mag = magnitude(v1,v2)

print(mag)
```

```python
#making arrays of positions and velocities

MW_pos = np.array([MW_data['x'], MW_data['y'], MW_data['z']])
M31_pos = np.array([M31_data['x'], M31_data['y'], M31_data['z']])
M33_pos = np.array([M33_data['x'], M33_data['y'], M33_data['z']])

MW_vel = np.array([MW_data['vx'], MW_data['vy'], MW_data['vz']])
M31_vel = np.array([M31_data['vx'], M31_data['vy'], M31_data['vz']])
M33_vel = np.array([M33_data['vx'], M33_data['vy'], M33_data['vz']])

#calling mag for MW and M31 to get pos and vel difference magnitudes 

MW_M31_pos = magnitude(MW_pos,M31_pos)
#print(MW_M31_pos) #test
MW_M31_vel = magnitude(MW_vel,M31_vel)

#calling mag for M33 and M31 to get pos and vel difference magnitudes 

M33_M31_pos = magnitude(M33_pos,M31_pos)
#print(M33_M31_pos) #test
M33_M31_vel = magnitude(M33_vel,M31_vel)
```

```python
# Plot the Orbit of the galaxies 
#################################

#for MW and M31
fig,ax = plt.subplots()
ax.plot(MW_data['t'],MW_M31_pos, label = "MW - M31 position", linewidth = "3") #t vs positional separation 
ax.plot(MW_data['t'],MW_M31_vel, linestyle = "--", color = "red", label = 'MW - M31 velocity', linewidth = "3") #t vs velocity separation
plt.xlabel("Time (GYR)")
plt.ylabel("Magnitude of Relative Separation")
plt.title("MW - M31 Pos and V Separations Over Time")
ax.legend()

#for M33 and M31
fig,ax = plt.subplots()
ax.plot(M33_data['t'],M33_M31_pos, label = "M33 - M31 position", linewidth = "3") #t vs positional separation 
ax.plot(M33_data['t'],M33_M31_vel, linestyle = "--", color = "red", label = 'M33 - M31 velocity', linewidth = "3") #t vs velocity separation
plt.xlabel("Time (GYR)")
plt.ylabel("Magnitude of Relative Separation")
plt.title("M33 - M31 Pos and V Separations Over Time")
ax.legend()

#pos and vel are plotted over each other to make compare more easily. 
```
